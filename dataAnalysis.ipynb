```python
import matplotlib.pyplot as plt
import seaborn as sns
import io



# Create histogram for Age distribution
plt.figure(figsize=(8, 6))
sns.histplot(dfC['Age'].dropna(), bins=40, kde=True, color='blue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Calculate average age for survivors and non-survivors
avg_age = dfC.groupby('Survived')['Age'].mean()
print("Average age by survival status:")
print(f"Survivors: {avg_age[1]:.2f} years")
print(f"Non-survivors: {avg_age[0]:.2f} years")
```

# Explore the relationship between 'Fare' and 'Survived' by grouping data and calculating average fare for survivors and non-survivors, then create a box plot

```python

# Calculate average fare for survivors and non-survivors
avg_fare = dfC.groupby('Survived')['Fare'].mean()
print("Average fare by survival status:")
print(f"Survivors: {avg_fare[1]:.2f}")
print(f"Non-survivors: {avg_fare[0]:.2f}")

# Create box plot for Fare by Survived
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=dfC, palette='viridis')
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare')
plt.show()
```

# Investigate gender-based survival by creating a crosstab of 'Gender' and 'Survived', and compute the survival percentage for males and females.

```python
# Calculate IQR for Fare
Q1 = dfC['Fare'].quantile(0.25)
Q3 = dfC['Fare'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 3 * IQR

# Identify outliers
outliers = dfC[dfC['Fare'] > upper_bound][['PassengerId', 'Fare']]
print("Passengers with Fare outliers (Fare > Q3 + 3*IQR):")
print(outliers)
```

# Analyze family size by creating a new column 'FamilySize' = SibSp + Parch + 1, then plot the survival rate against family size.

```python

# Create FamilySize column
dfC['FamilySize'] = dfC['SibSp'] + dfC['Parch'] + 1

# Calculate survival rate by FamilySize
survival_rate = dfC.groupby('FamilySize')['Survived'].mean() * 100

# Print survival rate
print("Survival rate by FamilySize (%):")
print(survival_rate)

# Create bar plot for survival rate by FamilySize
plt.figure(figsize=(8, 6))
sns.barplot(x=survival_rate.index, y=survival_rate.values, palette='viridis')
plt.title('Survival Rate by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate (%)')
plt.show()
```

# Examine embarkation points by counting passengers from each 'Embarked' location, and create a pie chart showing the proportion, along with survival rates per port

```python
# Count passengers by Embarked
embarked_counts = dfC['Embarked'].value_counts()
print("Passenger count by Embarked:")
print(embarked_counts)

# Calculate survival rate by Embarked
survival_rate = dfC.groupby('Embarked')['Survived'].mean() * 100
print("\nSurvival rate by Embarked (%):")
print(survival_rate)

# Create pie chart for Embarked proportions
plt.figure(figsize=(8, 6))
plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', colors=sns.color_palette('viridis'))
plt.title('Proportion of Passengers by Embarkation Point')
plt.show()
```

# Correlate numerical features by computing a correlation matrix for 'Age', 'Fare', 'SibSp', 'Parch', 'Pclass', and 'Survived', and visualize it as a heatmap.

```python
# Select numerical columns
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch', 'Pclass', 'Survived']
correlation_matrix = dfC[numerical_cols].corr()

# Print correlation matrix
print("Correlation matrix for numerical features:")
print(correlation_matrix)

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

# Handle missing 'Age' values by imputing them with the median age per 'Pclass' and 'Gender', then compare the age distribution before and after imputation.

```python
# Plot age distribution before imputation
plt.figure(figsize=(8, 6))
sns.histplot(dfC['Age'].dropna(), bins=30, kde=True, color='blue', label='Before Imputation')
plt.title('Age Distribution Before and After Imputation')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Impute missing Age values with median per Pclass and Gender
dfC['Age_imputed'] = dfC.groupby(['Pclass', 'Gender'])['Age'].transform(lambda x: x.fillna(x.median()))

# Plot age distribution after imputation
sns.histplot(dfC['Age_imputed'], bins=30, kde=True, color='orange', label='After Imputation', alpha=0.5)
plt.legend()
plt.show()

# Compare summary statistics
print("Age statistics before imputation:")
print(dfC['Age'].describe())
print("\nAge statistics after imputation:")
print(dfC['Age_imputed'].describe())
```

# Explore title extraction from 'Name' column (e.g., Mr., Mrs.), create a new 'Title' column, and analyze survival rates by title

```python

# Extract Title from Name
dfC['Title'] = dfC['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Calculate survival rate by Title
survival_rate = dfC.groupby('Title')['Survived'].mean() * 100
print("Survival rate by Title (%):")
print(survival_rate.sort_values(ascending=False))
```

# Investigate the impact of cabin deck (first letter of 'Cabin') on survival by extracting 'Deck', handling missing cabins, and plotting survival rates per deck

```python

# Extract Deck from Cabin (first letter)
dfC['Deck'] = dfC['Cabin'].str[0]

# Handle missing cabins by assigning 'Unknown'
dfC['Deck'] = dfC['Deck'].fillna('Unknown')

# Calculate survival rate by Deck
survival_rate = dfC.groupby('Deck')['Survived'].mean() * 100
print("Survival rate by Deck (%):")
print(survival_rate)

# Create bar plot for survival rate by Deck
plt.figure(figsize=(8, 6))
sns.barplot(x=survival_rate.index, y=survival_rate.values, palette='viridis')
plt.title('Survival Rate by Deck')
plt.xlabel('Deck')
plt.ylabel('Survival Rate (%)')
plt.show()
```

# Perform a multivariate analysis by creating a pairplot for 'Age', 'Fare', 'Pclass', and 'Survived' colored by 'Gender' to identify patterns.

```python

# Select columns for pairplot
pairplot_cols = ['Age', 'Fare', 'Pclass', 'Survived']

# Create pairplot
sns.pairplot(dfC[pairplot_cols].dropna(), hue='Survived', palette='viridis', diag_kind='kde')
plt.suptitle('Pairplot of Age, Fare, Pclass, and Survived', y=1.02)
plt.show()
```
